The NumPy library contains multidimensional array and matrix data structures. It provides ndarray, a homogeneous n-dimensional array object, with methods to efficiently operate on it.

## What’s the difference between a Python list and a NumPy array?
NumPy gives you an enormous range of fast and efficient ways of creating arrays and manipulating numerical data inside them. While a Python list can contain different data types within a single list, all of the elements in a NumPy array should be homogeneous. The mathematical operations that are meant to be performed on arrays would be extremely inefficient if the arrays weren’t homogeneous.



```python
x =[2,4,6,8,10]
z= np.array(x)
print(z.dtype)

print((x*2), "list multiplication: \n        multiplies the list increase the item in list equal to multiplied value\n")
print((z*2), "array multiplication: element-wise operation")
```

```python
a = [[x, x, x], [x, x, x]]
a=np.array(a)
print(a)
a.ndim
```

## Why use NumPy?
    NumPy arrays are faster and more compact than Python lists. An array consumes less memory and is convenient to use. NumPy uses much less memory to store data and it provides a mechanism of specifying the data types. This allows the code to be optimized even further.

### Numpy is fast 

- because of vectorization ( the absence of explicit looping , indexing) in the code
- all of the code are optimized ` looping is behind the scene` and  pre-compiled C code
    - vectorized code is more concise and easier to read
    - fewer lines of code generally means fewer bugs
    - the code more closely resembles standard mathematical notation (making it easier, typically, to correctly code mathematical constructs)
    -  vectorization results in more “Pythonic” code. Without vectorization, our code would be littered with inefficient and difficult to read for loops.


```python
A, B, A1= 5, 2, [2,3]
A1 = [2,3]
a, B1, Aa1= np.array(A), np.array(B), np.array(A1)

print('List_sum = ', A+B, '              Numpy_array_sum = ', a+B1)
print('List_sub = ', A-B, '              Numpy_array_sub = ', a-B1) # 'List_sub = ', A-B: cannot do action with list
print('List_mul = ', A1*B, '   Numpy_array_mul = ', Aa1*B)   # List_mul = ', A*B : TypeError: can't multiply sequence by non-int of type 'list
print('List_div = ', A/B, '            Numpy_array_div = ', Aa1/B) # List_div = ', A/B : TypeError: unsupported operand type(s) for /: 'list' and 'list
print('numpy array does operation in itemwise')
```

```python
A, B = [2,5,7], [3,5,9]

Aa, Ba = np.array(A),np.array(B)


print('List_sum = ', A+B, '   Numpy_array_sum = ', Aa+Ba)
print('\n                                  Numpy_array_sub = ', Aa-Ba) # 'List_sub = ', A-B: cannot do action with list
print('\n                                  Numpy_array_mul = ', Aa*Ba)   # List_mul = ', A*B : TypeError: can't multiply sequence by non-int of type 'list
print('\n                                  Numpy_array_div = ', Aa/Ba) # List_div = ', A/B : TypeError: unsupported operand type(s) for /: 'list' and 'list
print('numpy array does operation in itemwise')
```

## filtering techniques

1. Where

```python
# where """Return elements chosen from x or y depending on condition."""

arr = np.array([1,2,4,6,7,8,9,])
y =np.where(arr < 8)
print(y)

z= np.extract(y,arr)
print(z)
```

```python
# Divide by 2
print(b)
div = b[b%2 == 0]
print(div)
```

```python
# conditions:
c = b[(b<7) & (b>3)]
print(c)
```

```python
# Importing libraries
import numpy as np               # ref: https://numpy.org/doc/
import matplotlib.pyplot as plt  # ref: https://matplotlib.org/stable/index.html
import pandas as pd              # ref: https://pandas.pydata.org/docs/
import sklearn                   # ref: https://scikit-learn.org/stable/user_guide.html

# Importing dataset
dataset = pd.read_csv('50_Startups.csv')
print(dataset.head())
X = dataset.iloc[:, :-2].values  # 
print("\nif you wish to include catagorical variable 'State', you will need OneHotEncoder from sklearn.preprocessing. \nFor simplification, we will ignore 'State' variable and leave until we reach to 'Machine Learning - Multiple Linear Regression' topic\n")
y = dataset.iloc[:, -1].values
print(dataset.head(), "\n******************** shape = ", dataset.shape, ' ********************************\n')
print(X[:5])
print('******************** data type : ', type(X), '  ***********\n')
print(y[:5])
print('******************** data type : ', type(y), '  ***********\n')
```

```python
# splitting data for training and test purpose

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Training
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)  # np.set_printoptions(precision=2)
print("Training accuracy score : ", regressor.score(X_train,y_train)*100)
print("Test accuracy score : ", regressor.score(X_test,y_test)*100)
print()
```

```python
print("Predicted Profit : ", y_pred)
print("Actual profit : ", y_test)
plt.scatter(y_pred,y_test,s=5)
plt.legend()
plt.xlabel("Predicted Profit")
plt.ylabel("Actual Profit")
plt.title('Linear Regression')
```
